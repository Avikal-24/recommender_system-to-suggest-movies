## Merging the 2 datasets to create our final dataset

```python
dataset=pd.merge( rating_details , movie_details , on='movieId' )
dataset.head(30)

# This is because movieId was the common attribute. 
# hence we merged ON THE BASIS OF movieId
```

```python
dataset.tail(10)
```

Changing timestamp column to correct date-time format

```python
#import datetime
#from datetime import datetime,date

dataset['datetime'] = pd.to_datetime(dataset['timestamp'])
dataset.tail(10)
```

```python
dataset.nunique()
```

```python
dataset.describe()
```

## Getting average rating

```python
a=dataset
a=a.groupby('title')['rating'].mean()
a

# groupby function groups the data according to the columns
# now 'a' shows 1 movie in ONLY 1 row with mean ratings of different users
```

```python
a.sort_values(ascending=False)
# sorted movies according to their ratings
```

```python
b=dataset.groupby('title')['rating'].count()
b.tail()

# it shows how many people have rated the particular movie
```

```python
new_record = pd.DataFrame()
new_record['Average_ratings'] = a
new_record['Count of total ratings'] = b
new_record.head(10)
```

```python
new_record.columns
```

```python
plt.figure(figsize=(12,12))
new_record['Count of total ratings'].hist(bins=200)
```

Hence most of the movies have less than 50 ratings

```python
plt.figure(figsize=(12,12))
new_record['Average_ratings'].hist(bins=200)
```

Hence majority of the movies have a rating between 3 and 4

```python
plt.figure(figsize=(15,15))
sns.jointplot(x='Average_ratings' , y='Count of total ratings' , data=new_record , alpha=0.5)
```

Hence we conclude that movies with higher AVERAGE RATINGS have higher COUNT OF RATERS 

AND VICE VERSA

```python
movie_matrix= dataset.pivot_table( index='userId' , columns='title' , values='rating' )
movie_matrix.tail(10)

# basically shows which user has rated what all movies and by how much
```

## Now we'll predict similar movies to "eXistenZ (1999)"

```python
existenz_ratings=movie_matrix['eXistenZ (1999)']
existenz_ratings
```

```python
similar_movie= movie_matrix.corrwith(existenz_ratings)
similar_movie
```

```python
correlation_metrics= dataset.corr()
fig=plt.figure(figsize=(15,15))
sns.heatmap(correlation_metrics,  annot=True, vmax=0.9, square=True )
plt.show()
```

```python
existenz_corr = pd.DataFrame( similar_movie , columns=['Correlation'] )
existenz_corr
```

This marks what all movies are related to Existenz and by how much
